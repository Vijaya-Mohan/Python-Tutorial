In Python, you make a comment by starting the line with #.
Anything after # is ignored by Python 

Comments are helpful because they:

Explain what your code does – so you don’t forget later.
Help others understand your code – useful if someone else reads it.
Make complex code easier to follow – especially when learning.
Let you leave notes for yourself – like reminders or ideas.

Think of comments as sticky notes inside your code.
They don't change how your programme runs – they just make it easier to understand.

Python does not have true multi-line comments. But  multi-line comment is usually a docstring. Text written inside triple quotes (""".... """) is called a docstring.

```python
# This is a comment in Python
print("This is a comment") # This is also a comment
"""
This is a comment
written in
more than just one line
"""
print("This is a multi-line comment")
```

Indentation:
Python uses indentation (spaces at the start of a line) to show which lines of code belong together.
You normally use 4 spaces.
If the spacing is wrong, Python can’t understand your code and shows errors like “unexpected indent” or “expected an indented block”.
Just keep the same number of spaces for each block of code.

```python
if 10>5:
    print("10 is greater than 5")
```

Rules for Python Variables


A variable name must start with a letter or an underscore (_)
A variable name cannot start with a number
A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
A variable name is case-sensitive (age, Age and AGE are three different variables)
A variable name cannot be a Python keyword


```python
myvar = "Mat"
my_var = "Mat"
_my_var = "Mat"
myVar = "Mat"
MYVAR = "Mat"
myvar2 = "Mat"
print(myvar, my_var, _my_var, myVar, MYVAR, myvar2)
```


Python automatically works out what type of value it is, so you don’t need to tell Python automatically decides the type of a variable based on the value you give it.
This means you do not need to tell Python the type – Python works it out for you,it whether it’s a number or text.

The type() function tells you how Python sees the value stored in a variable.

```python
x = 10
print("Value of x:",x)
x1 = 2.5
print("Value of x1:",x1)
y = 'Eva'
print("Value of y:",y)
z = "Evan" # Other way to declare a string
print("Value of z:",z) 

is_logged_in = True
game_over = False


print("Type of x:",type(x))
print("Type of x1:",type(x1))
print("Type of y:",type(y))
print("Type of z:",type(z))
print("Type of is_logged_in:",type(is_logged_in))
print("Type of game_over:",type(game_over))
```

What is type conversion? <br>
Type conversion means changing one data type into another.<br>
For example, changing text into a number, or a number into text.<br>
Python allows this using built‑in functions.

Why do we need type conversion? <br>
Python reads input as text (string), even when we want numbers.<br>
To do calculations, we must convert the data type.

```python
x = "5"
print("Type of x:",type(x))

# Convert string to integer
y = int(x) 
print("Type of y:",type(y)) 
```
